```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from hrvanalysis import remove_outliers, remove_ectopic_beats, interpolate_nan_values
import json

pd.set_option("display.max_columns", 100)
```

```python
xqrs_full = pd.read_csv('../to_csv/cleaned_files SWT XQRS/xqrs_v2_cleaned.csv').drop(['Unnamed: 0'], axis=1)
xqrs_full.head(10)
```

```python
cols = xqrs_full.columns.tolist()
cols = cols[:6] + cols[10:-5]
cols
```

```python
xqrs = xqrs_full[cols]
xqrs.head()
```

```python
xqrs['Date'] = xqrs.Session.str[-10:]
xqrs['Date'] = xqrs.Date.str.replace('_', '/')
xqrs['Date'] =  pd.to_datetime(xqrs['Date'], format='%Y/%m/%d').dt.strftime('%d/%m/%Y')

xqrs.Session = xqrs.Session.str[:4]

cols = xqrs.columns.tolist()
cols = cols[:5] + ['Date'] + cols[5:-1]

xqrs = xqrs[cols]

xqrs.head()
```

```python
xqrs['Dossier Patient'] = xqrs['Dossier Patient'].astype(str)
xqrs['Patient'] = xqrs['Patient'].astype(str)
```

```python
xqrs['Key'] = xqrs.Set + '_' + xqrs['Categorie Montage'] + '_' + xqrs['Patient'] + '_' + xqrs.Session + '_' + xqrs['File N°']
```

```python
cols = xqrs.columns.tolist()
cols = cols[-1:] + cols[7:-1]
xqrs_lite = xqrs[cols]
xqrs_lite
```

```python
xqrs_lite.shape
```

```python
xqrs_lite = xqrs_lite.dropna(subset=['label'])
xqrs_lite.shape
```

```python
xqrs_lite = xqrs_lite.replace([np.inf, -np.inf], np.nan)
```

```python
xqrs_lite_na = xqrs_lite.groupby(xqrs_lite['Key']).transform(lambda x: x.fillna(x.mean()))
```

```python
xqrs_lite_na['Key'] = xqrs_lite.Key.values
```

```python
cols = xqrs_lite_na.columns.tolist()
cols = cols[-1:] + cols[:-1]
xqrs_lite_na = xqrs_lite_na[cols]
xqrs_lite_na
```
