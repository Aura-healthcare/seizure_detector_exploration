```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", 50)
```

```python
pan = pd.read_csv('../to_csv/pan.csv').drop('Unnamed: 0', axis=1)
pan.head()
```

```python
pan['Date'] = pan.Session.str[-10:]
pan['Date'] = pan.Date.str.replace('_', '/')
pan['Date'] =  pd.to_datetime(pan['Date'], format='%Y/%m/%d').dt.strftime('%d/%m/%Y')

pan.Session = pan.Session.str[:4]

cols = pan.columns.tolist()
cols = cols[:5] + ['Date'] + cols[5:-1]

pan = pan[cols]

pan.head()
```

```python
pan.describe(include='all')
```

```python
pan['Dossier Patient'] = pan['Dossier Patient'].astype(str)
pan['Patient'] = pan['Patient'].astype(str)
```

```python
pan.describe(include='all')
```

```python
percentage = pan[['Patient', 'label']].groupby(['Patient']).agg(['sum', 'count'])
```

```python
percentage.columns
```

```python
percentage[('label', 'percent')] = (percentage[('label',   'sum')] / percentage[('label',   'count')]) * 100
```
