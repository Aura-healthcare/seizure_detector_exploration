```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", 50)
```

```python
swt = pd.read_csv('../to_csv/swt.csv').drop('Unnamed: 0', axis=1)
swt.head()
```

```python
swt['Date'] = swt.Session.str[-10:]
swt['Date'] = swt.Date.str.replace('_', '/')
swt['Date'] =  pd.to_datetime(swt['Date'], format='%Y/%m/%d').dt.strftime('%d/%m/%Y')

swt.Session = swt.Session.str[:4]

cols = swt.columns.tolist()
cols = cols[:5] + ['Date'] + cols[5:-1]

swt = swt[cols]

swt.head()
```

```python
swt.describe(include='all')
```

```python
swt['Dossier Patient'] = swt['Dossier Patient'].astype(str)
swt['Patient'] = swt['Patient'].astype(str)
```

```python
swt.describe(include='all')
```

```python
percentage = swt[['Patient', 'label']].groupby(['Patient']).agg(['sum', 'count'])
```

```python
percentage.columns
```

```python
percentage[('label', 'percent')] = (percentage[('label',   'sum')] / percentage[('label',   'count')]) * 100
```
